```python
import numpy as np
```

```python
mit_train = np.genfromtxt("mitbih_train.csv",delimiter=",")
```

```python
mit_train_x_complete = mit_train[:,:-1]
mit_train_y_complete = mit_train[:,-1]
```

```python
np.random.seed(6)
train_samp = np.random.choice(mit_train_x_complete.shape[0],replace=False,size=80000)
mit_train_x = mit_train_x_complete[train_samp]
mit_train_y = mit_train_y_complete[train_samp]
#mit_train_x = mit_train_x_complete[:]
#mit_train_y = mit_train_y_complete[:]

mit_val_x = mit_train_x_complete[np.setdiff1d(np.arange(mit_train_x_complete.shape[0]),train_samp)]
mit_val_y = mit_train_y_complete[np.setdiff1d(np.arange(mit_train_x_complete.shape[0]),train_samp)]
```

```python
mit_test = np.genfromtxt("mitbih_test.csv",delimiter=",")
```

```python
mit_test_x = mit_test[:,:-1]
mit_test_y = mit_test[:,-1]
mit_test_x.shape,mit_test_y.shape
```

```python
import sys
sys.path.append("../pytorch-softdtw-cuda")
import soft_dtw_cuda
```

```python
sdtw = soft_dtw_cuda.SoftDTW(use_cuda=True, gamma=1e-12)
```

```python
import torch
```

```python
device = 'cuda'
```

```python
batch_size = 10000
x_pieces = []
i=0
while batch_size*i < mit_train_x_complete.shape[0]:
    x_pieces.append(torch.tensor(mit_train_x_complete[batch_size*i:batch_size*(i+1)]).unsqueeze(2).to(device))
    i += 1
    print(x_pieces[-1].shape)
```

```python
preds = []
for i in range(mit_test_x.shape[0]):
    y = torch.tensor(mit_test_x[i:i+1]).unsqueeze(2).repeat(batch_size,1,1).to(device)
    y_lastpiece = torch.tensor(mit_test_x[i:i+1]).unsqueeze(2).repeat(x_pieces[-1].shape[0],1,1).to(device)
    pred = []
    print("start",i)
    for j,x_piece in enumerate(x_pieces[:-1]):
        #print(j,x_piece.shape)
        loss = sdtw(x_piece,y)
        pred += list(loss.detach().cpu().numpy())
    loss = sdtw(x_pieces[-1],y_lastpiece)
    pred += list(loss.detach().cpu().numpy())
    preds.append(pred)
    print(i)
```

```python
preds_npy = np.array(preds)
```

```python
from collections import Counter
pred_3nn = []
for i in range(mit_test_y.shape[0]):
    class_counts = Counter()
    for j in range(3):
        class_counts[mit_train_y_complete[sorted_preds[i][j]]] += 1
    most_com = class_counts.most_common()
    pred_3nn.append(most_com[0][0])
```

```python
from collections import Counter
pred_5nn = []
for i in range(mit_test_y.shape[0]):
    class_counts = Counter()
    for j in range(5):
        class_counts[mit_train_y_complete[sorted_preds[i][j]]] += 1
    most_com = class_counts.most_common()
    pred_5nn.append(most_com[0][0])
```
